```python
import numpy as np
import seaborn as sns
# Sample dataset
data = [114950, 158000, 230500, 387000, 389950, 479000, 488800, 529000, 575000, 639000, 659000, 1095000, 5500000]
data.sort()
print(data)
# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data,25) # (n-1) * percentile + 1
Q3 = np.percentile(data,75)
IQR = Q3 - Q1
# Determine bounds
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
# Identify outliers
outliers = [x for x in data if x < low_lim or x > up_lim]
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Limit: {low_lim}, Upper Limit: {up_lim}")
print(f"Outliers: {outliers}")
```

```python
import pandas as pd
import numpy as np

data = [114950, 158000, 230500, 387000, 389950, 479000, 488800, 529000, 575000, 639000, 659000, 1095000, 5500000]

# Convert your list to a Pandas Series for easy statistical calculations
s = pd.Series(data)

# Calculate all the statistics
mean_val = s.mean()
median_val = s.median()
mode_val = s.mode()[0] #.mode() can return multiple, so [0] gets the first one
Q1 = s.quantile(0.25) # (n-1) * percentile + 1
Q2 = s.quantile(0.50) # This is the same as the median
Q3 = s.quantile(0.75)
Q4 = s.max() # The maximum value is equivalent to the 100th percentile
IQR = Q3 - Q1

# Let's put these into a descriptive DataFrame, similar to what you were trying
descriptive_stats = pd.DataFrame({
    'Statistic': [
        'Mean', 'Median', 'Mode', 'Q1 (25%)', 'Q2 (50%)', 'Q3 (75%)', 'Q4 (100%)', 'IQR'
    ],
    'Value': [
        mean_val, median_val, mode_val, Q1, Q2, Q3, Q4, IQR
    ]
})

print("--- Descriptive Statistics for Your Data ---")
print(descriptive_stats.set_index('Statistic'))
```

# Manual Method

My Manual Method (Method 1: Exclusive Median)
This is a very common method for manual calculation:

Find the Median (Q2).

Divide the data into two halves based on the median.
If n is odd, exclude the median from both halves.
If n is even, split exactly in the middle.

Q1 is the median of the lower half.
Q3 is the median of the upper half.

Applying it to your data:
data =
n = 13

Median (Q2): 7th value = 488800.

Lower half (excluding median): ``
Median of lower half (n=6): Average of 3rd and 4th values: (230500 + 387000) / 2 = 308750. So, Q1 = 308750.

Upper half (excluding median): ``
Median of upper half (n=6): Average of 3rd and 4th values: (639000 + 659000) / 2 = 649000. So, Q3 = 649000.
